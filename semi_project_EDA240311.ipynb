```python
selected_columns = selected_columns.drop(1)
selected_columns.head(10)
```

```python
selected_columns = selected_columns.reset_index(drop=True)
```

```python
selected_columns.head(10)
```

```python
selected_columns['119안전센터 1개소 당 담당인구'] = selected_columns['119안전센터 1개소 당 담당인구'].astype(int)
selected_columns_sorted = selected_columns.sort_values(by='119안전센터 1개소 당 담당인구', ascending=False)
selected_columns_sorted.head(10)
```
